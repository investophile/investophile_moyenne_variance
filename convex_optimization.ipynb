```python
# Set up the data
x_array = np.array(data)

n = x_array.shape[1]

# Compute the logarithmic returns
log_returns = np.diff(np.log(x_array),axis=0)
```

```python
# Compute the expected values vector
mu = np.mean(log_returns,axis=0)

# Compute the covariance matrix
Sigma = np.cov(log_returns.T)

print("The Expected Value vector\n\n",mu)
print("\n")
print("First 7 rows & columns of the Covariance Matrix\n\n",np.round(Sigma[:7,:7],7))
```

```python
def mean_variance(mu: np.ndarray, sigma_inv: np.ndarray) -> np.ndarray:
    """
    Calculates the optimal portfolio weights for mean-variance optimization.

    Parameters:
    - mu (np.ndarray): Expected returns of individual assets.
    - sigma_inv (np.ndarray): Inverse of the covariance matrix of asset returns.

    Returns:
    - np.ndarray: Optimal portfolio weights that minimize portfolio variance for a given level of expected return.
    """

    # Calculate the portfolio weights using mean-variance optimization formula
    w0 = np.dot(sigma_inv, mu)

    # Normalize the weights to ensure they sum up to 1
    return w0 / np.matmul(np.abs(w0.T), np.ones(len(w0)))
```

```python
# Compute the weights of the optimal portfolio
w = mean_variance(mu,np.linalg.inv(Sigma))

print("Standard Mean Variance Optimal Portfolio\n\n",np.round(w,4))
```

```python
import cvxpy as cvx
def long_only_mean_variance(Sigma: np.ndarray, mu: np.ndarray, risk_tolerance: float) -> np.ndarray:
    """
    Markowitz Long Only optimal portfolio.

    Parameters:
    - Sigma (np.ndarray): Covariance matrix of asset returns.
    - mu (np.ndarray): Expected returns of individual assets.
    - riskTolerance (float): Risk tolerance parameter for the mean-variance optimization.

    Returns:
    - np.ndarray: Optimal long-only portfolio weights based on mean-variance optimization.
    """

    # Total number of assets
    n = len(mu)

    # Variable vector of length n for optimization
    z = cvx.Variable(n)

    # The expected value for a portfolio of weights w
    ret = mu @ z

    # The variance for a portfolio of weights w
    risk = cvx.quad_form((w + z), Sigma)

    # The objective function to minimize
    objective = cvx.Minimize(risk - risk_tolerance * ret)

    # The (linear) constraints
    constraints = [np.ones((1, n)) @ (w + z) <= 1, (w + z) >= 0]

    # Optimization problem
    prob = cvx.Problem(objective, constraints)

    # Solve the optimization problem
    prob.solve(verbose=False)

    # Return the updated portfolio weights
    return w + np.asarray(z.value)
```

```python
# Investor's risk tolerance
risk_tolerance = 0.1

# Compute the weights of the optimal portfolio
w = long_only_mean_variance(Sigma,mu,risk_tolerance)

print("Long Only Mean Variance Optimal Portfolio\n\n",np.round(w,4))
```
